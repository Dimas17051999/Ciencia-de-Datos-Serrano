```python
import requests
from bs4 import BeautifulSoup

pagina = requests.get("https://en.wikipedia.org/wiki/List_of_current_National_Football_League_stadiums")
print(pagina.content)
```

```python
soup = BeautifulSoup(pagina.content, 'html.parser')
print(soup.prettify())
```

```python
body = soup.find('body')
print(body.get_text())
```

```python
th = soup.find_all('th')
for i in th:
    print(i.get_text())
```

```python
names = soup.find_all('th', attrs={'scope':'row'})
count = 0
for i in names: 
    count = count + 1
    print(i.get_text())
    if count ==30: 
        break
```
